```python
import numpy as np

# Tumhara data: area (sqft) aur price (lakhs)
area   = np.array([500,  800,  1000, 1200, 1500])
price  = np.array([25,   38,   50,   58,   72  ])

# Manual linear relationship: price ≈ area * 0.05 - 0.5
# (yeh "model" hai — ek formula jo pattern pakadta hai)
def predict_price(area_sqft):
    return area_sqft * 0.05 - 0.5

# Predict karo
new_areas = np.array([600, 900, 1100])
predictions = predict_price(new_areas)
print("Predicted prices (lakhs):", predictions)

# Error calculate karo (actual vs predicted)
predicted_on_known = predict_price(area)
errors = price - predicted_on_known
print("Errors:", errors.round(2))
print("Mean error:", np.mean(np.abs(errors)).round(2), "lakhs")

# Visualization
import matplotlib.pyplot as plt

plt.scatter(area, price, color='blue', label='Actual data', zorder=5)
x_line = np.linspace(400, 1600, 100)
plt.plot(x_line, predict_price(x_line), color='red', label='Our model')
plt.xlabel("Area (sqft)")
plt.ylabel("Price (lakhs)")
plt.title("Our first 'model' — manual!")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

kg=np.array([1,2,3,4,5])  #imput
price=np.array([20,40,60,80,100])   #output

#  x value have to convert in 2d array
# (sample,feature)

# shape of x (sample,feature)
print("1d dhape of inpute",kg)
x=kg.reshape(-1,1)
print("2d shape of input data",x)

model=LinearRegression()

model.fit(x,price)

new_data=np.array([[6]])

result=model.predict(new_data)

print("price of 6 kg is ", result[0],"rupees")

# erro_cal=

import matplotlib.pyplot as plt

plt.scatter(kg, price, color='blue', label='Actual data', zorder=5)
# x_line = np.array([6,7,8,3,2])
plt.plot(x, result(x), color='red', label='Our model')
plt.xlabel("Area (sqft)")
plt.ylabel("Price (lakhs)")
plt.title("Our first 'model' — manual!")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import pandas as pd

data = {
    'Age':    [22, np.nan, 23, 28, 150, 25, 27],
    'Salary': [35000, 52000, np.nan, 61000, 29000, 44000, 58000],
    'City':   ['Bhopal', 'Delhi', 'Mumbai', 'Bhopal', 'Delhi', 'Mumbai', 'Delhi'],
    'Gender': ['Female', 'Male', 'Female', 'M', 'Female', 'Male', 'Female'],
    'Bought': [1, 0, 1, 1, 0, 1, 0]  # yeh predict karna hai
}

df = pd.DataFrame(data)
print(df)
print("\nShape:", df.shape)
```
